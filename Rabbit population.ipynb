```python
def populationGrowth(time, rate, k, init):

  # base case
  if time == 0:
    return init # initial population

  lastPt = populationGrowth(time - 1, rate, k, init) # get population from previous year
  pt = lastPt + (rate*lastPt * (1-(lastPt/k))) # new population

  print("Population in year ", time, " is ", pt)
  return pt
```

```python
def main(menu):
  while menu:
    print("POPULATION GROWTH \n")

    time = 0
    rate = -1
    k = -1
    init = -1

    # input validation
    while time < 1:
      time = int(input("Enter expected time: "))
      if time < 1:
        print("Invalid time. Can't compute no time nor negative time.")

    while rate < 0 or rate > 1:
      rate = float(input("Enter growth rate: "))
      if rate < 0 or rate > 1:
        print("Invalid growth rate. Rate input must be a number between 0 and 1.")

    while k < 0:
      k = int(input("Enter carrying capacity: "))
      if k < 2:
        print("Invalid carrying capacity. Carrying capacity must be greater than 0.")

    while init < 2:
      init = int(input("Enter initial population: "))
      if init < 2:
        print("Invalid initial population. Initial population must be at least 2.")

    print("\n\n")
    populationGrowth(time, rate, k, init)

    print("\n\n")
    answer = str(input("Would you like to calculate for a new environment? [Y/n]"))

    print("\n\n")

    if answer != 'Y' or answer != 'y':
      menu = False
      if answer != 'N' or answer != 'n':
        print("Invalid input.")
      print("Ending program...")
```

```python
main(True)
```
