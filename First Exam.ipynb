```python
# Importar bibliotecas
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, roc_auc_score, precision_recall_curve
)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset desde un archivo CSV
archivo_csv = '/data.csv'
df = pd.read_csv(archivo_csv)

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar un modelo de regresión logística
model = LogisticRegression()
model.fit(X_train, y_train)

# Predecir en el conjunto de prueba
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probabilidades para la curva ROC

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_proba)

# Mostrar métricas
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Sensibilidad (Recall): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"AUC: {auc:.2f}")


```

## Excercise 3 Searches
Searching a communication network graph. Problem description:

We have a directed graph that represents a communication network
between servers. Each node represents a server, and each edge
represents a connection with an associated cost (transmission time
in milliseconds). The objective is to find the optimal route from a
start server to a destination server using search algorithms.


Grafo:
* A B 10
* A C 15
* B D 12
* C D 10
* D E 5

Heurísticas:
* A E 20
* B E 15
* C E 10
* D E 5
* E E 0

```python
graph = {
    'A': [('B', 10), ('C', 15)],
    'B': [('D', 12)],
    'C': [('D', 10)],
    'D': [('E', 5)],
    'E': []
}

heuristics = {
    'A': 20,
    'B': 15,
    'C': 10,
    'D': 5,
    'E': 0
}
```

```python
import heapq

def a_star(grafo, origen, destino, heuristica):
    lista_abierta = [(0 + heuristica[origen], 0, origen)]

    lista_cerrada = set()
    costos = {origen: 0}
    padres = {origen: None}

    while lista_abierta:
        # Pop the node with the lowest f(n)
        f_actual, g_actual, nodo_actual = heapq.heappop(lista_abierta)

        # If the current node is the destination, terminate and reconstruct the path
        if nodo_actual == destino:
            path = []
            while nodo_actual is not None:
                path.append(nodo_actual)
                nodo_actual = padres[nodo_actual]
            return path[::-1]  # Return the path from origin to destination

        # Move the current node to the closed list (visited node)
        lista_cerrada.add(nodo_actual)

        for vecino, costo_arco in grafo.get(nodo_actual, []):
            if vecino in lista_cerrada:
                continue  # Skip if the neighbor is already visited

            nuevo_g = g_actual + costo_arco

            if vecino not in costos or nuevo_g < costos[vecino]:
                costos[vecino] = nuevo_g
                f_vecino = nuevo_g + heuristica[vecino]  # f(n) = g(n) + h(n)

                # Add the neighbor to the open list with its f(n) value
                heapq.heappush(lista_abierta, (f_vecino, nuevo_g, vecino))
                padres[vecino] = nodo_actual

    # Open list is empty and destination is not found
    return "No hay camino disponible"

# Solution to the excercise
path = a_star(graph, 'A', 'E', heuristics)
print("Camino encontrado:", path)
```
